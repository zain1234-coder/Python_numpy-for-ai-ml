# np.sort

Return a sorted copy of an numpy array

```python
import numpy as np 

a = np .random.randint(1,100,15)
a2 = np .random.randint(1,100,24).reshape(6,4)
print(a,"\n")

print(a2)
```

```python
np.sort(a)
```

```python
np.sort(a2)  # sort row wise
```

```python
np.sort(a2,axis =0 ) # sorting column wise
```

# set funtions
 - np.union    **1d**()   
 - np.intersect   **1d**()   
 - np.setdiff     **1d**()   
 - np.setxor      **1d**()   
 - np.isin        **1d** ()   


```python
m = np.array([1,2,3,4,5,6,6])
n = np.array([4,2,6,80,98,76,5])

print("union",np.union1d(m,n))

print("intersection",np.intersect1d(m,n))
print("Set difference", np.setdiff1d(m,n))
print("setxor",np.setdiff1d(m,n))
print(np.isin(m,4))
```

```python
m[np.isin(m,3)]
```
